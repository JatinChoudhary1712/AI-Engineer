```python
import time

def task(name):
    print(f"{name} started")
    time.sleep(5)
    print(f"{name} finished")

start = time.time()

task("A")
task("B")

print("Total:", time.time() - start)
```

```python
import asyncio
import time

async def task(name):
    print(f"{name} started")

    await asyncio.sleep(5)

    print(f"{name} finished")

async def main():
    start = time.time()

    await asyncio.gather(
        task("A"),
        task("B")
    )

    print("Total:", time.time() - start)

await main()
```

```python
import asyncio

async def coffee():
    print("Making coffee")
    await asyncio.sleep(3)
    print("Coffee ready")

async def cookie():
    print("Baking cookie")
    await asyncio.sleep(2)
    print("Cookie ready")

async def main():
    await asyncio.gather(
        coffee(),
        cookie()
    )

await main()
```
