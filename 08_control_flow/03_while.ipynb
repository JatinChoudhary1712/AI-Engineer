while loop → runs while condition is True

```python
i = 0
while i < 3:
    print(i)
    i += 1
```

break → exits loop immediately

```python
for i in range(5):
    if i == 3:
        break
    print(i)   # 0,1,2
```

continue → skips current iteration

```python
for i in range(5):
    if i == 3:
        continue
    print(i)   # 0,1,2,4
```
