*args → multiple positional arguments

```python
def add(*args):
    print(args)

add(1, 2, 3)   # (1, 2, 3)
```

**kwargs → multiple keyword arguments

```python
def info(**kwargs):
    print(kwargs)

info(name="Jatin", age=22)
```

```python
def demo(*args, **kwargs):
    print(args)
    print(kwargs)

demo(1, 2, name="Jatin", age=22)
```
