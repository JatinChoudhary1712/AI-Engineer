* **Indexing** → Access element using position: `lst[0]` (first), `lst[-1]` (last)
* **Slicing** → Get sublist: `lst[start:end:step]` (end excluded)
* **append()** → Adds **single element** at end: `lst.append(x)`
* **extend()** → Adds **multiple elements** (iterable): `lst.extend([a,b])`
* **pop()** → Removes & returns element (default last): `lst.pop()` / `lst.pop(i)`
* **sort()** → Sorts list in-place: `lst.sort()` (use `reverse=True` for descending)
* **reverse()** → Reverses list in-place: `lst.reverse()`
