#### uniqueness : stores only unique values

```python
s = {1, 1, 2, 3}   # {1, 2, 3}
print(s)
```

#### union 

```python
s1 = {1,2,3 , 3 ,4}
s2 = {4,5,6}
s1 | s2 # s1.union(s2) 
```

#### Intersection

```python
s1 & s2 # s1.intersection(s2)
```

#### Difference

```python
s1 - s2 # s1.difference(s2)
```
