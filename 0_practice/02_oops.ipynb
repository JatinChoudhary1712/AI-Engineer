```python
class Student():
    def __init__(self , name , age , marks):
        self.name = name
        self.age = age 
        self.marks = marks
    def display_info(self):
        return f"Name : {self.name}, Age : {self.age},  Marks : {self.marks}"
```

```python
s1 = Student("Jatin" , 23 , 100)
```

```python
s1.display_info()
```

```python
s2 = Student("Kajal" , 26 , 100)
```

```python
s2.display_info()
```

-Polymorphism

```python
class Person():
    def __init__(self , name):
        self.name = name 
    
    def show_details(self):
        return f"The person name is {self.name}"
```

```python
class Employee(Person): 
    def __init__(self, name , salary):
        super().__init__(name)
        self.salary = salary
    
    def show_details(self):
        return f"The name of Person : {self.name} and Salary is {self.salary}"
```

```python
p1 = Employee("Jatin" , "12 Lakh")
```

```python
p1.show_details()
```

```python
p2 = Person("Rahul")
```

```python
print(p2.show_details())
```
