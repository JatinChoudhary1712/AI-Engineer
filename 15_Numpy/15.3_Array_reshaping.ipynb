1. reshape() : Used to change the structure of an array.

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(2,3)
```

```python
new_arr
```

2. flatten()

Converts multidimensional array into 1D array.

```python
arr = np.array([[1,2,3], [4,5,6]])

arr.flatten()
```

3. ravel()

Also converts to 1D.

```python
print(arr.ravel())
```
