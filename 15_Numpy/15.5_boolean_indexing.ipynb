1. Boolean Indexing

Means:

filtering array using conditions

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr > 25)
```

```python
# Basic Syntax 

result = np.where(arr > 25, 1,0 ) 
```

```python
result
```

```python
arr = np.array([-1,5,-3,8])

print(np.where(arr < 0, 0, arr))
```
