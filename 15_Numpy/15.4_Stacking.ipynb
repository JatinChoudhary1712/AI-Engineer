1. np.stack()

- General stacking function.

- It joins arrays along a new axis.

```python
import numpy as np 

a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
result = np.stack((a,b))
```

```python
result
```

Axis

-  Default:

axis=0

Meaning:
stack row-wise.

axis=1

Meaning : coloumns wise joining


```python
np.stack((a,b) , axis=1)
```
