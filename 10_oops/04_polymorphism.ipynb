```python
class Car:
    def start(self, model):
        return f"Car : {model}"

class ElectricCar(Car):
    def start(self , model):
        return f"Car (EV) : {model}"

c1 = Car()
c2 = ElectricCar()

print(c1.start("Mustang"))
print(c2.start("Tesla"))
```
