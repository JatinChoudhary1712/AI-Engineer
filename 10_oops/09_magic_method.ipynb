def (__str__)

- 👉 Used when you do print(object)

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student name is {self.name}"

s = Student("Jatin")
print(s)
```

2. __repr__ → Developer/debug view

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"

s = Student("Jatin")
print(s)
```

3. __len__ → length behavior

```python
class Basket:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

b = Basket(["apple", "banana", "mango"])
print(len(b))
```

```python
class Basket:
    def __init__(self, items):
        self.items = items

    def __getitem__(self, index):
        return self.items[index]

b = Basket(["apple", "banana", "mango"])
print(b[1])
```
