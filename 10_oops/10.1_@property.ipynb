```python
class Student:
    def __init__(self, age):
        self._age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value < 0:
            print("Invalid age")
        else:
            self._age = value

s1 = Student(20)

s1.age = 25
print(s1.age)
```
