```python
class Car: 
    def __init__(self ,brand , model):
        self.brand = brand 
        self.model = model
    def full_name(self):
        print(f"Hi my car brand and name is {self.brand} and {self.model}")
```

```python
my_car = Car("Toyota" , "corolla")
print(my_car.brand , my_car.model)
```

```python
my_car.full_name()
```
