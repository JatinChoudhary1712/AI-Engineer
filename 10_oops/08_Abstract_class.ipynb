```python
from abc import ABC , abstractmethod

class Loader(ABC):
    @abstractmethod 
    def load(self):
        pass 
```

```python
class PDFLoader(Loader):
    def load(self):
        print("Loading PDF")
```

```python
class WebLoader(Loader):
    def load(self):
        print("Loading Website")
```

```python
from abc import ABC, abstractmethod

class DataSource(ABC):

    @abstractmethod
    def fetch_data(self):
        pass


class APIData(DataSource):
    def fetch_data(self):
        return "Data from API"


class DBData(DataSource):
    def fetch_data(self):
        return "Data from Database"
```

```python
def process_data(source: DataSource):
    print(source.fetch_data())
```
