```python
class Books():
    def __init__(self , books):
        self.books = books
    def __str__(self):
        return f"The list of book is {self.books}"
    def __len__(self):
        return len(self.books)
    def __getitem__(self, item):
        return self.books[item]
```

```python
b = Books(["Hero" , "Villian" , "Heroine" , "Rimjhiim"])
```
