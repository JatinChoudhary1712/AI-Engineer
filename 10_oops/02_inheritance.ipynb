```python
class Car():
    def __init__(self , brand , model):
        self.brand = brand 
        self.model = model
```

```python
new_car = Car("Suzuki" , "Fronx")
```

- Inheritance

```python
class ElectricCar(Car):
    def __init__(self, brand, model , battery_size):
        super().__init__(brand, model)
        self.battery_size = battery_size
        
    def full_name(self):
        return (f"The Car company {self.brand} ,{self.model} and battery size {self.battery_size}")
```

```python
my_tesla = ElectricCar("Tesla" , "Model S" , "85Kwh")
```

```python
my_tesla.brand
```

```python
print(my_tesla.full_name())
```
