# 🔹 Core Python Types (1-Liners)

- **int** → whole numbers like `10`
- **float** → decimal numbers like `10.5`
- **str** → text data like `"hello"`
- **bool** → logical values `True` or `False`
- **None** → represents absence of a value (not 0 or empty)

```python
print(type(10))
```

```python
print(type(10.5))
```

```python
print(type("hello"))
```

```python
print(bool("False"))
```

```python
x = 1
bool(x > 2)
```
