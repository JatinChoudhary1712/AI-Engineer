1. count()

Infinite counter.

```python
from itertools import count

counter = count(1)


print(next(counter))
print(next(counter))
print(next(counter))
```

2. cycle()

Repeats forever.

```python
from itertools import cycle 

colors = cycle(["red" , "blue"])

print(next(colors))
print(next(colors))
print(next(colors))
```

3. repeat()

Repeats same value.

```python
from itertools import repeat

for i in repeat("AI", 3):
    print(i)
```

4. MOST IMPORTANT → chain()

Combines iterables.

Very useful in:

- merging datasets
- preprocessing pipelines

```python
from itertools import chain 

a = [1,2]
b = [3,4]

for i in chain(a,b):
    print(i)
```

5. SUPER IMPORTANT → islice()

Like slicing but memory efficient.

```python
from itertools import islice

nums = range(1000000)

print(list(islice(nums, 12)))
```
