```python
nums = [1,2,3,4,5]

result = []

for i in nums : 
    result.append(i)
    
print(result)
```

Vectorization

Using optimized low-level operations instead of Python loop.

Usually done with:

- NumPy
- Pandas

```python
import numpy as np 
nums = np.array([1,2,3,4,5])

result = nums*2

print(result)
```
