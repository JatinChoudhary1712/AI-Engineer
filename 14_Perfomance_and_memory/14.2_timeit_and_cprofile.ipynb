timeit

Measures execution time.

```python
import timeit

code = """
sum(range(100))
"""

print(timeit.timeit(code, number=1000))
```

cProfile 

Help us understand where our program is slow

```python
import cProfile

def test():
    total = 0
    for i in range(100000):
        total += i

cProfile.run("test()")
```
