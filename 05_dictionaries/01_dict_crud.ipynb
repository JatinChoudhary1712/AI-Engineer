### CRUD (Create, Read, Update, Delete)
- Create → d = {"a":1, "b":2}
- Read → d["a"] → 1
- Update → d["a"] = 10
- Add new → d["c"] = 3
- Delete → del d["a"] or d.pop("a")

#### Create

```python
d = {"a":1 , "b":2}
```
