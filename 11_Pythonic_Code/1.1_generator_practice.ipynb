```python
nums = range(1, 11) 

def even():
    for i in nums : 
        if i%2 == 0 :
            yield i 
            
result = even()

for x  in result:
    print(x)
```

```python
nums = range(1, 11) 

def square():
    for i in nums:
        if i**2 > 20 : 
            yield i**2
            
            
result = square()
```

```python
for x in result:
    print(x)
```

```python
words = ["cat", "dog", "python"] 


def reversed_words():
    for i in words: 
        yield i[::-1]
```

```python
result = reversed_words()
```

```python
for x in result:
    print(x)
```
