Pythomic code mean writing the python code in python way 
- Cleaner, shorter, readable
- Using built-in tools instead of unnecessary loops

```python
# non pythonic 
squares = []
for i in range(5):
    squares.append(i*i)

print(squares)
```

```python
# pythonic 
squares = [i*i for i in range(5)]
print(squares)
```

```python
names = ["jatin", "rahul", "aman"]

capitalized = [name.capitalize() for name in names]

print(capitalized)
```

2. Dictionary Comprehensions

- Creates dictionaries in one line.

```python
nums = [1, 2, 3, 4]

square_dict = {x: x*x for x in nums}

print(square_dict)
```

```python
students = ["Jatin", "Rahul", "Aman"]

lengths = {name: len(name) for name in students}

print(lengths)
```

3. Set Comprehensions

- Creates sets.

```python
nums = [1, 2, 2, 3, 3, 4]

unique = {x for x in nums}

print(unique)
```
