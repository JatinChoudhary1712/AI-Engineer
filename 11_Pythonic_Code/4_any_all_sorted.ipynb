```python
marks = [32, 28, 91, 40]

result = any(mark > 105 for mark in marks)

print(result)
```

```python
nums = [1, 2, 3]

print(all(nums))
```

```python
passwords = ["abc123", "hello99", "python77"]

result = all(len(p) >= 6 for p in passwords)

print(result)
```

```python
nums = [2, 4, 6, 8]

print(any(n % 2 == 1 for n in nums))
print(all(n % 2 == 0 for n in nums))
```
