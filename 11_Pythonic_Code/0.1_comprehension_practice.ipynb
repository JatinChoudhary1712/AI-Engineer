Dict Comprehension

---

```python
nums = [5, -2, 7, -1, 8]

result = [x if x > 0 else 0 for x in nums]
```

```python
result
```

```python
nums = [1, 2, 3, 4] 

square = {x : x**2 for x in nums}

square
```

```python
nums = [1, 2, 3, 4]
cube = {x:x**3 for x in nums}
```

```python
cube
```

```python
words = ["apple", "banana", "cat"]

count = {x : len(x) for x in words}
```

```python
count
```

```python
words = ["python", "java", "sql"]

{x : x.upper() for x in words}
```

```python
nums = [1, 2, 3, 4, 5, 6] 

even_square = {x : x**2 for x in nums if x%2 ==0}
```

```python
even_square
```

```python
data = {
    "a": 1,
    "b": 2,
    "c": 3
} 

swapped = {v:k for k ,v in data.items() }
```

```python
swapped
```

```python
marks = {
    "Rahul": 80,
    "Aman": 32,
    "Priya": 67
}

result = {
    k: "Pass" if v > 32 else "Fail"
    for k, v in marks.items()
}
```

```python
result
```

```python
nums = [2, 5, 8, 11, 14] 

result = {
    k : "Even" if k%2 == 0 else  "Odd" 
    for k in nums
}
```

```python
result
```

```python
marks = {
    "Rahul": 80,
    "Aman": 32,
    "Priya": 67,
    "Karan": 45
} 
```

```python
result = {
    k:v  for k,v in marks.items() if v > 50
}
```

```python
result
```
