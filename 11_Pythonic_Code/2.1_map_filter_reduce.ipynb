1. map()

- Used to:

-- apply a function to every item of iterable

- syntax

map(function , iterable)

```python
nums = [1, 2, 3, 4]

result = map(lambda x: x*x, nums)

print(list(result))
```

```python
words = ["python", "java", "sql"] 

result = map(lambda x : x.upper() , words)
```

```python
print(list(result))
```

2. filter()

Used to:

filter/select values based on condition

Syntax : filter(function, iterable)

```python
nums = [1, 2, 3, 4, 5, 6] 

nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)

print(list(result))
```

---

```python
words = ["cat", "python", "java", "AI"] 

result = filter(lambda x : len(x) > 4 , words)

print(list(result))
```

```python
from functools import reduce 

nums = [1, 2, 3, 4]

result = reduce(lambda a,b : a +b , nums)
```
