| Syntax       | Meaning                  |
| ------------ | ------------------------ |
| `a, b = lst` | unpack values            |
| `*rest`      | collect remaining values |
| `_`          | ignore value             |


```python
nums = [10, 20] 

a,b = nums
```

```python
print(b)
```

```python
lst = [1, 2, 3, 4, 5] 

a , b , *rest = lst 

print(a)
print(b)
print(rest)
```

```python
nums = [100, 200, 300, 400, 500] 
a , *b , c = nums
```

```python
values = [5, 10, 15] 

a, _, c = values

print(a)
print(c)
```

```python
point = (4, 9) 

x,y = point
```

```python
pairs = [(1, 2), (3, 4), (5, 6)] 


for a, b in pairs :
    print(a, b)
```
