File handling

---

Reading

```python
with open("data.txt", "r") as file:
    content = file.read()

print(content)
```

---

Writing

```python
with open("notes.txt" , "w") as file:
    file.write("Hello Jatin Choudhary")
```

---

Append

```python
with open("notes.txt" , "a") as file:
    file.write("\n Hello Kajal")
```
